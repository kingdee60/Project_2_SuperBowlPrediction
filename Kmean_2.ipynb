```python
# Import the required libraries and dependencies
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
season_2021_2011= pd.read_csv("2021_2011_regular_season.csv", 
                          index_col=0,
                          parse_dates = True, 
                          infer_datetime_format = True
)
```

```python
season_2021_2011
```

```python
# season_2021_2011.dtypes
```

```python
season_2021_2011["Home_code"]= season_2021_2011["Home"].astype("category").cat.codes
```

```python
season_2021_2011["Road_code"]= season_2021_2011["Road"].astype("category").cat.codes
```

```python
df_season = season_2021_2011[["W", "L"]]
```

```python
# df_season
```

```python
# Start by importing the K-means algorithm
from sklearn.cluster import KMeans
```

```python
# Create and initialize the K-means model instance for 2 clusters
model = KMeans(n_clusters=2, random_state=1)

# Print the model
model
```

```python
# predictors = [season_2021_2011["W"], 
#               season_2021_2011["L"], 
#               season_2021_2011["PCT"],
#               season_2021_2011["Home_code"], 
#               season_2021_2011["Road_code"]]
```

```python
# Fit the data to the instance of the model
model.fit(df_season)
```

```python
team_ratings = model.predict(df_season)
```

```python
print(team_ratings)
```

```python
team_rating_predictions_df= df_season.copy()
```

```python
team_rating_predictions_df["team_ratings"] = team_ratings
```

```python
team_rating_predictions_df.head()
```

```python
team_rating_predictions_df.hvplot.scatter(
    x="W", 
    y="L", 
    by="team_ratings")
```
